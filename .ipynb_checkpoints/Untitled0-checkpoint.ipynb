```python
%matplotlib inline
import pylab
import numpy as np
from IPython.display import display, Math, Latex
```

```python
data=[]
with open('lista.txt','r') as file:
    for line in file.readlines():
        data.append(np.loadtxt(line.split()[0]))
```

```python
E_max=[]
for i in data:
    pylab.plot(i[:,0],i[:,1])
    E_max.append(max(i[:,1]))
```

```python
pylab.plot(np.linspace(0,len(E_max)*np.pi*np.pi/20,len(E_max)),E_max)
pylab.xlabel(r'$\omega$',size=20)
pylab.ylabel(r'$<\epsilon>_{max}$',size=20)
```
